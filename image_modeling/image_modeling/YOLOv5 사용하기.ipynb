```python
# 모든 파일의 경로 가져오기
# 하나의 변수에 모든 이미지 값을 넣어야함 = X
# 하나의 변수에 모든 라벨링 값을 넣어야함 = y
# 각 파일의 경로를 한번에 모음 > 학습에 사용

from glob import glob

train_img_list = glob('/content/dataset/train/images/*.jpg')
```

```python
test_img_list = glob('/content/dataset/test/images/*.jpg')
val_img_list = glob('/content/dataset/valid/images/*.jpg')
```

```python
# 이미지 경로들을 하나의 파일에 담아줌
# train.txt, val.txt
# w : write(쓰기)
with open('/content/dataset/train.txt' , 'w') as f:
  f.write('\n'.join(train_img_list) + '\n')

with open('/content/dataset/val.txt' , 'w') as f:
  f.write('\n'.join(val_img_list) + '\n')
```

```python
# data.yaml파일 수정하기
import yaml

with open('/content/dataset/data.yaml','r') as f:
  data = yaml.safe_load(f)

print(data)
```

```python
data['train'] = '/content/dataset/train.txt'
data['val'] = '/content/dataset/val.txt'
```

```python
# data를 yaml파일에 덮어쓰기
# data.yaml파일에는 모든 파일의 경로가 적혀있다. > 학습시에 data.yaml파일만 제공
with open('/content/dataset/data.yaml','w') as f:
  yaml.dump(data,f)
```
